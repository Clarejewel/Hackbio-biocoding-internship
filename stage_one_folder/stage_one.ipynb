```python
# Write a function for translating DNA to protein
# First define and assign the codons in amino acids using a dictionary. I will be working with the first 10 amino acids
# Dictionary of amino acids and their codons
amino_acids = {
    "Alanine": ["GCT", "GCC", "GCA", "GCG"],
    "Arginine": ["CGT", "CGC", "CGA", "CGG", "AGA", "AGG"],
    "Asparagine": ["AAT", "AAC"],
    "Aspartic Acid": ["GAT", "GAC"],
    "Cysteine": ["TGT", "TGC"],
    "Glutamic Acid": ["GAA", "GAG"],
    "Glutamine": ["CAA", "CAG"],
    "Glycine": ["GGT", "GGC", "GGA", "GGG"],
    "Histidine": ["CAT", "CAC"],
    "Isoleucine": ["ATT", "ATC", "ATA"]

}

# To confirm my dictionary was created, print all the Alanine codons
# print(amino_acids)
# print(amino_acids["Alanine"])

for amino_acid in amino_acids:
    # prints the amino acid and its codons
    print(amino_acid, amino_acids[amino_acid])

# Next, I will write a function that translates DNA to protein.
# The function will take a DNA sequence as input and return the corresponding protein sequence.
# The function will loop through the DNA sequence in steps of 3 to read the DNA sequence in codons (3 nucleotides at a time).
# It will then check if the codon exists as a key in the dictionary, amino_acids, and add the corresponding amino acid to the protein sequence.
# Finally, it will return the translated protein sequence.
#
#  Because codons are made of 3 nucleotides, we loop through the DNA sequence in steps of 3.as_integer_ratio
# Read the DNA sequence in codons (3 nucleotides at a time)


def protein_translation(dna_sequences):
    # protein sequence will be the final output. Assigned an empty string
    protein_sequence = ""

    # generates numbers starting from 0 up to len(dna_sequence), increasing by 3 each time.#i represents the starting index of each codon.
    for i in range(0, len(dna_sequences), 3):
        codon = dna_sequences[i:i+3]  # Extracts 3-letter codon at each step.

        # loops through the dictionary, amino_acids, to find matching codon.
        for amino_acid, codons in amino_acids.items():
            if codon in codons:  # checks if the codon exists as a key in the dictionary, amino_acids.
                # adds the amino acid name to the protein sequence.
                protein_sequence += amino_acid + " "
    return protein_sequence  # returns the translated protein sequence.


# Test the function with a sample DNA sequence
# Sample DNA sequence to test my function
dna_sequences = (
    "CACATATCCATTCCGACCAAACCATGATAAAGTATATGCCACGTCACTATTCATCGTCGATTAGTTTTTA")

# calling the function
protein_translation(dna_sequences)
```

```python
# Write a function that simulates and generates a logistic population growth curve. Your function should include 2 extra parameters that randomize the length of the lag phase and the exponential phase [See population curve here - https://media.springernature.com/full/springer-static/image/art%3A10.1038%2Fsrep15159/MediaObjects/41598_2015_Article_BFsrep15159_Fig1_HTML.jpg).
# Most living populations follow a logistic population growth. Therefore, your growth curve can be: Population Size vs Time, Cell density vs Time, OD vs Time, CFU vs Time, etc

# Solution

# Working with population size vs Time, the logistic growth curve formula is given by:
# P(t) = K/1+e^-r(t-tmid))

# where:
# P(t) = the population at time
# K = the carrying capacity (maximum population).
# r = the growth rate (how quickly the population grows).
# tmid = the midpoint or inflection point (the time when the growth rate switches from accelerating to decelerating).

# the function will be

import math  # importing the math module to use the exponential function


# t is time, K is the carrying capacity, r is the growth rate, t_mid is the midpoint
def growth_curve_generator(t, K, r, t_mid):
    # the logistic growth curve formula
    answer = K/(1 + math.exp(-r*(t-t_mid)))
    return answer  # returns the answer


# Testing the function with random values for the parameters where t = 10, K = 100,r = 0.1, t_mid = 5


new_population = growth_curve_generator(t=10, K=100, r=0.1, t_mid=5)
print(f"The population size at time t = 10 is {new_population}")

# calling the function
# The output
# 62.245933120185455
# The output is the population size at time t = 10 with the given parameters.
# The population size increases from 0 to 62.2 as time increases from 0 to 10, following the logistic growth curve formula.
```

```python
# Using your function, generate a dataframe with 100 different growth curves
# 100 different growth curves would have different time points with different values of K, r, and t_mid.

# 1. FIrst create a list of time points from say, 0 to 25 using the numpy linspace function. It would be a timeline (0  25) with 100 time points.
# Something to note: "np.linspace(start, stop, num)" - Creates evenly spaced numbers between start and stop. The num parameter determines how many numbers to generate.

# importing the matplotlib library to plot the growth curves
import matplotlib.pyplot as plt
import pandas as pd  # importing the pandas library to create a dataframe
# importing the uniform distribution from scipy.stats
from scipy.stats import uniform
import numpy as np
# creates 100 different time points from 0 to 20.
diff_time_points = np.linspace(0, 25, 100)
print(diff_time_points)

# 2. The function will use random values for K, r, and t_mid for each growth curve.
# The random values will be generated using the scipy.stats.uniform.rvs().
# #This function generates random variables from a uniform distribution.
# random_values = uniform.rvs(loc=low, scale=high-low, size=n) where loc is the lower bound, scale is the upper bound, and size is the number of random variables to generate.

# generates 100 random values for K between 10 and 150.
K_values = uniform.rvs(loc=10, scale=150, size=100)
# generates 100 random values for r between 0.01 and 0.25.
r_values = uniform.rvs(loc=0.01, scale=0.25, size=100)
# generates 100 random values for t_mid between 5 and 50.
t_mid_values = uniform.rvs(loc=5, scale=50, size=100)

# 3. Then compute the 100 different growth curves using the function
# The function will loop through the 100 time points and calculate the population size at each time point using the logistic growth curve formula.
# The population size at each time point will be stored in a dictionary.

# empty dictionary to store the population size at each time point.
growth_curve_data = {}

# Generating the 100 different growth curves. The function will store the population size at each time point in a list.

for i in range(100):  # loops through the 100 time points.
    K = K_values[i]  # gets the random value for K.
    r = r_values[i]  # gets the random value for r.
    t_mid = t_mid_values[i]  # gets the random value for t_mid.
    # empty list to store the population size at each time point.
    population_size = []
    for t in diff_time_points:  # loops through the time points.
        # calculates the population size at each time point.
        population = growth_curve_generator(t, K, r, t_mid)
        # appends the population size to the list.
        population_size.append(population)
    # stores the population size at each time point in the dictionary.
    growth_curve_data[f"Growth Curve {i+1}"] = population_size

# 4. Finally, the function will create a dataframe with the time points and the population size at each time point.
# creates a dataframe with the population size at each time point.
df = pd.DataFrame(growth_curve_data, index=diff_time_points)
print(df)

# Let's visualize some of the growth curves using a line plot.
# Plotting the first 5 growth curves
df.iloc[:, :5].plot()  # plots the first 5 growth curves.
plt.xlabel("Time")  # x-axis label
plt.ylabel("Population Size")  # y-axis label
plt.title("Logistic Population Growth Curves")  # title of the plot
plt.show()  # displays the plot
```

```python
# Write a function for determining the time to reach 80% of the maximum growth; usually the carrying capacity
# Time to reach 80% of maximum growth is P(t) = 0.8K.

# Imput this in this formula "P(t) = K/1+e^-r(t-tmid))" and make t the subject of the formula

# t = t_mid - log(1/0.8) - 1)/r where 0.8 is the percentage (i.e 80%)

# Now create the function
def time_to_80_percent(K, r, t_mid):
    # calculates the time to reach 80% of the maximum growth.
    t_80 = t_mid - (math.log(1/0.8) - 1)/r
    return t_80  # returns the time to reach 80% of the maximum growth.


# Testing the function with random values for K =100, r = 0.1, t_mid = 5
time_80_percent = time_to_80_percent(K=100, r=0.1, t_mid=5)
print(
    f"The time to reach 80% of the maximum growth is {time_80_percent} units of time.")
```

```python
# Finally, write a function for calculating the hamming distance between your Slack username and twitter/X handle (synthesize if you dont have one). Feel free to pad it with extra words if they are not of the same length.
# The Hamming distance between two strings of equal length is the number of positions at which the corresponding characters are different.

Slack_username = "JudithMba"
Twitter_handle = "clarejudy"

# First define the function for calculating the Hamming distance between two strings.


def hamming_distance(s1, s2):
    distance = 0  # initialize the distance to 0.
    for i in range(len(s1)):  # loops through the characters in the strings.
        if s1[i] != s2[i]:  # checks if the characters are not equal.
            distance += 1  # increments the distance by 1.
    return distance  # returns the Hamming distance.


# Testing the function with the Slack username and Twitter handle.
hamming_dist = hamming_distance(Slack_username, Twitter_handle)
print(
    f"The Hamming distance between the Slack username and Twitter handle is {hamming_dist}.")
# The output: The Hamming distance between the Slack username and Twitter handle is 9.
```
