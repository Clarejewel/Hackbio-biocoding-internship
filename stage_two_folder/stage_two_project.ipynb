```python
# Using your function from last stage, determine the time to reach the carrying capacity for each strain/mutant
# I'll create a function that covers/contains the plotting logic developed earlier. This function will take the DataFrame and the sample mapping as inputs.to calculate the time to reach carrying capacity.

# Import necessary libraries
import urllib.request
import ssl
import pandas as pd
import matplotlib.pyplot as plt
import io


def analyze_growth_curves(data_df, mapping, k_threshold_percent=0.95, show_plots=True):
    """
    Plots growth curves and calculates the time to reach carrying capacity (K).

    Args:
        data_df (pd.DataFrame): DataFrame with 'time' column and OD data columns (e.g., 'A1', 'A2',...).
        mapping (dict): Dictionary mapping column names (e.g., 'A1') to strain details
                        (e.g., {'strain': 'Strain1', 'type': 'WT', 'rep': 'Rep1', 'cond': 'A'}).
        k_threshold_percent (float): The percentage of max OD to define reaching K (default: 0.95).
        show_plots (bool): Whether to display the growth curve plots (default: True).

    Returns:
        pd.DataFrame: A DataFrame containing the analysis results for each replicate:
                      Strain, Type, Replicate, Condition, Max_OD (K), Time_to_K_hours.
                      Returns None if input data is invalid.
    """
    # Import numpy here to ensure it's available within the function scope
    import numpy as np

    if 'time' not in data_df.columns:
        print("Error: 'time' column not found in DataFrame.")
        return None

    # Convert time to hours (assuming input is minutes)
    time_points_hours = data_df['time'] / 60.0

    # Get unique strain names
    unique_strains = sorted(
        list(set(details['strain'] for details in mapping.values())))
    print(f"Analyzing strains: {unique_strains}")

    results = []  # To store analysis results

    # --- Plotting (Optional) ---
    if show_plots:
        n_strains = len(unique_strains)
        fig, axes = plt.subplots(n_strains, 1, figsize=(
            10, n_strains * 4.5), sharex=True, squeeze=False)
        fig.suptitle('Growth Curves (OD600 vs. Time)', fontsize=16, y=1.0)

        # Define colors and styles
        wt_color = 'blue'
        mut_color = 'red'
        wt_style = '-'
        mut_style = '--'
        marker_size = 4
        alpha_val = 0.7

    # --- Analysis Loop ---
    for i, strain_name in enumerate(unique_strains):
        if show_plots:
            ax = axes[i, 0]  # Current subplot axis

        # Find columns corresponding to the current strain
        strain_cols_map = {col: details for col, details in mapping.items(
        ) if details['strain'] == strain_name}

        plotted_wt_legend = False
        plotted_mut_legend = False

        for col, details in strain_cols_map.items():
            if col not in data_df.columns:
                print(
                    f"Warning: Column '{col}' defined in mapping but not found in data. Skipping.")
                continue

            od_values = data_df[col]
            strain_type = details['type']
            replicate = details['rep']
            condition = details['cond']  # Keep track of condition (A, B, C)

            # Calculate K (Max OD)
            max_od = od_values.max()

            # Calculate threshold OD
            threshold_od = max_od * k_threshold_percent

            # Find the first time index where OD >= threshold
            # Note: idxmax() finds the first occurrence of the maximum value,
            # here we find the first occurrence >= threshold
            time_indices_reaching_k = time_points_hours[od_values >= threshold_od]

            time_to_k = time_indices_reaching_k.iloc[0] if not time_indices_reaching_k.empty else np.nan

            # Store results
            results.append({
                'Strain': strain_name,
                'Type': strain_type,
                'Replicate': replicate,
                'Condition': condition,
                'Max_OD (K)': max_od,
                f'Threshold_OD ({k_threshold_percent*100}%)': threshold_od,
                'Time_to_K_hours': time_to_k
            })

            # Plotting logic (if enabled)
            if show_plots:
                is_wt = (strain_type == 'WT')
                color = wt_color if is_wt else mut_color
                style = wt_style if is_wt else mut_style
                marker = 'o' if is_wt else 'x'
                # Add legend label only once per type per plot
                label = None
                if is_wt and not plotted_wt_legend:
                    label = f'{strain_name} WT'
                    plotted_wt_legend = True
                elif not is_wt and not plotted_mut_legend:
                    label = f'{strain_name} MUT'
                    plotted_mut_legend = True

                ax.plot(time_points_hours, od_values, marker=marker, linestyle=style, color=color,
                        label=label, markersize=marker_size, alpha=alpha_val)

                # Optionally mark the time_to_k point
                if not np.isnan(time_to_k):
                    ax.scatter(time_to_k, od_values[time_points_hours == time_to_k].iloc[0],
                               color='black', s=50, zorder=5, label=f'_nolegend_K_{col}')  # Mark point

        # Finalize plot for the current strain
        if show_plots:
            ax.set_title(f'Strain: {strain_name}')
            ax.set_ylabel('OD600')
            ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
            ax.grid(True, linestyle='--', alpha=0.6)

    # Final plot adjustments
    if show_plots:
        axes[-1, 0].set_xlabel('Time (hours)')
        plt.tight_layout(rect=[0, 0, 0.9, 0.98])  # Adjust layout
        plt.show()

    # Convert results to DataFrame
    results_df = pd.DataFrame(results)
    # Reorder columns for clarity
    column_order = ['Strain', 'Condition', 'Type', 'Replicate', 'Max_OD (K)',
                    f'Threshold_OD ({k_threshold_percent*100}%)', 'Time_to_K_hours']
    results_df = results_df[column_order]

    return results_df


# --- Data Loading and Mapping ---
# Load the dataset from the link


url = 'https://raw.githubusercontent.com/HackBio-Internship/2025_project_collection/refs/heads/main/Python/Dataset/mcgc.tsv'

# Disable SSL verification
ssl._create_default_https_context = ssl._create_unverified_context

# Download the file content
try:
    response = urllib.request.urlopen(url)
    data_string = response.read().decode('utf-8')
except urllib.error.URLError as e:
    print(f"Failed to fetch data from the URL. Error: {e}")
    exit()

# Load the data into a DataFrame
data = pd.read_csv(io.StringIO(data_string), sep='\t')

# Look at the first few rows
print(data.head())

# Define the mapping based on the data description
mapping = {
    'A1': {'strain': 'Strain1', 'type': 'WT', 'rep': 'Rep1', 'cond': 'A'},
    'A2': {'strain': 'Strain1', 'type': 'MUT', 'rep': 'Rep1', 'cond': 'A'},
    # ... (rest of the mapping as defined earlier)
}

# Ensure data_string is defined before using it
if 'data_string' not in locals():
    try:
        response = urllib.request.urlopen(url)
        data_string = response.read().decode('utf-8')
    except urllib.error.URLError as e:
        print(f"Failed to fetch data from the URL. Error: {e}")
        exit()

# Load the data
try:
    data = pd.read_csv(io.StringIO(data_string), sep='\t')
except pd.errors.ParserError as e:
    print(f"Error parsing data. Please check the format. Details:\n{e}")
    exit()
except Exception as e:
    print(f"An unexpected error occurred during data loading: {e}")
    exit()

# --- Run the analysis ---
# Call the function with the loaded data and mapping, getting the results DataFrame and showing plots
analysis_results = analyze_growth_curves(
    data, mapping, k_threshold_percent=0.95, show_plots=True)

# Determine the time to reach the carrying capacity for each strain/mutant
# Display the results table
if analysis_results is not None:
    print("\n--- Time to Reach 95% Carrying Capacity (K) ---")
    # Use try-except for tabulate in case it's still not installed
    try:
        # Ensure tabulate is installed: pip install tabulate
        print(analysis_results.round(3).to_markdown(
            index=False, numalign="left", stralign="left"))
    except ImportError:
        print("Info: 'tabulate' package not installed. Printing DataFrame without markdown formatting.")
        print(analysis_results.round(3))
    except Exception as e:
        print(f"An error occurred during results display: {e}")
```

```python
# --- Function Definition ---

def analyze_growth_curves(data_df, mapping, k_threshold_percent=0.95, show_plots=True):
    """
    Plots growth curves and calculates the time to reach carrying capacity (K).

    Args:
        data_df (pd.DataFrame): DataFrame with 'time' column and OD data columns (e.g., 'A1', 'A2',...).
        mapping (dict): Dictionary mapping column names (e.g., 'A1') to strain details
                        (e.g., {'strain': 'Strain1', 'type': 'WT', 'rep': 'Rep1', 'cond': 'A'}).
        k_threshold_percent (float): The percentage of max OD to define reaching K (default: 0.95).
        show_plots (bool): Whether to display the growth curve plots (default: True).

    Returns:
        pd.DataFrame: A DataFrame containing the analysis results for each replicate:
                      Strain, Type, Replicate, Condition, Max_OD (K), Time_to_K_hours.
                      Returns None if input data is invalid.
    """
    # Import numpy here to ensure it's available within the function scope
    import numpy as np

    if 'time' not in data_df.columns:
        print("Error: 'time' column not found in DataFrame.")
        return None

    # Convert time to hours (assuming input is minutes)
    time_points_hours = data_df['time'] / 60.0

    # Get unique strain names
    unique_strains = sorted(
        list(set(details['strain'] for details in mapping.values())))
    # Only print if plotting is enabled, to avoid duplicate messages
    if show_plots:
        print(f"Analyzing strains for plotting: {unique_strains}")

    results = []  # To store analysis results

    # --- Plotting Setup (Optional) ---
    fig = None
    axes = None
    if show_plots:
        n_strains = len(unique_strains)
        fig, axes = plt.subplots(n_strains, 1, figsize=(
            10, n_strains * 4.5), sharex=True, squeeze=False)
        fig.suptitle('Growth Curves (OD600 vs. Time)', fontsize=16, y=1.0)

        # Define colors and styles
        wt_color = 'blue'
        mut_color = 'red'
        wt_style = '-'
        mut_style = '--'
        marker_size = 4
        alpha_val = 0.7

    # --- Analysis Loop ---
    for i, strain_name in enumerate(unique_strains):
        ax = None  # Initialize ax
        if show_plots:
            ax = axes[i, 0]  # Current subplot axis

        # Find columns corresponding to the current strain
        strain_cols_map = {col: details for col, details in mapping.items(
        ) if details['strain'] == strain_name}

        plotted_wt_legend = False
        plotted_mut_legend = False

        for col, details in strain_cols_map.items():
            if col not in data_df.columns:
                print(
                    f"Warning: Column '{col}' defined in mapping but not found in data. Skipping.")
                continue

            od_values = data_df[col]
            strain_type = details['type']
            replicate = details['rep']
            condition = details['cond']  # Keep track of condition (A, B, C)

            # Calculate K (Max OD)
            max_od = od_values.max()

            # Calculate threshold OD
            threshold_od = max_od * k_threshold_percent

            # Find the first time index where OD >= threshold
            time_indices_reaching_k = time_points_hours[od_values >= threshold_od]

            # Use np.isnan() which is now guaranteed to be defined
            time_to_k = time_indices_reaching_k.iloc[0] if not time_indices_reaching_k.empty else np.nan

            # Store results
            results.append({
                'Strain': strain_name,
                'Type': strain_type,
                'Replicate': replicate,
                'Condition': condition,
                'Max_OD (K)': max_od,
                f'Threshold_OD ({k_threshold_percent*100}%)': threshold_od,
                'Time_to_K_hours': time_to_k
            })

            # Plotting logic (if enabled)
            if show_plots and ax is not None:
                is_wt = (strain_type == 'WT')
                color = wt_color if is_wt else mut_color
                style = wt_style if is_wt else mut_style
                marker = 'o' if is_wt else 'x'
                # Add legend label only once per type per plot
                label = None
                if is_wt and not plotted_wt_legend:
                    label = f'{strain_name} WT'
                    plotted_wt_legend = True
                elif not is_wt and not plotted_mut_legend:
                    label = f'{strain_name} MUT'
                    plotted_mut_legend = True

                ax.plot(time_points_hours, od_values, marker=marker, linestyle=style, color=color,
                        label=label, markersize=marker_size, alpha=alpha_val)

                # Optionally mark the time_to_k point
                # Use np.isnan() which is now guaranteed to be defined
                if not np.isnan(time_to_k):
                    # Ensure the index exists before trying to access it
                    matching_times = od_values[time_points_hours == time_to_k]
                    if not matching_times.empty:
                        ax.scatter(time_to_k, matching_times.iloc[0],
                                   color='black', s=50, zorder=5, label=f'_nolegend_K_{col}')  # Mark point

        # Finalize plot for the current strain
        if show_plots and ax is not None:
            ax.set_title(f'Strain: {strain_name}')
            ax.set_ylabel('OD600')
            # Only add legend if there are labels to show
            handles, labels = ax.get_legend_handles_labels()
            if labels:
                ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
            ax.grid(True, linestyle='--', alpha=0.6)

    # Final plot adjustments
    if show_plots and fig is not None:
        axes[-1, 0].set_xlabel('Time (hours)')
        plt.tight_layout(rect=[0, 0, 0.9, 0.98])  # Adjust layout
        plt.show()

    # Convert results to DataFrame
    results_df = pd.DataFrame(results)
    # Reorder columns for clarity
    column_order = ['Strain', 'Condition', 'Type', 'Replicate', 'Max_OD (K)',
                    f'Threshold_OD ({k_threshold_percent*100}%)', 'Time_to_K_hours']
    # Ensure columns exist before reordering
    results_df = results_df[[
        col for col in column_order if col in results_df.columns]]

    return results_df

# --- Data Loading and Mapping ---


# Pasted TSV data from the user
data_string = """
time	A1	A2	A3	A4	A5	A6	A7	A8	A9	A10	A11	A12	B1	B2	B3	B4	B5	B6	B7	B8	B9	B10	B11	B12	C1	C2	C3	C4	C5	C6	C7	C8	C9	C10	C11	C12
0	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002
15	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002
30	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002
45	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002
60	0.003	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.003	0.003	0.002	0.002	0.002	0.003	0.003	0.002	0.002	0.002	0.002	0.002	0.003	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002	0.002
75	0.003	0.002	0.003	0.002	0.002	0.003	0.002	0.003	0.003	0.003	0.003	0.003	0.002	0.003	0.002	0.003	0.003	0.003	0.002	0.003	0.003	0.003	0.003	0.003	0.002	0.002	0.003	0.002	0.002	0.002	0.002	0.003	0.002	0.002	0.003	0.003
90	0.003	0.002	0.003	0.002	0.003	0.003	0.003	0.003	0.003	0.003	0.003	0.003	0.002	0.003	0.002	0.003	0.003	0.003	0.002	0.003	0.003	0.003	0.003	0.003	0.002	0.003	0.003	0.002	0.002	0.002	0.002	0.003	0.003	0.002	0.003	0.003
105	0.004	0.003	0.003	0.002	0.003	0.003	0.003	0.003	0.003	0.003	0.004	0.004	0.002	0.003	0.002	0.004	0.004	0.003	0.003	0.004	0.003	0.003	0.004	0.003	0.002	0.003	0.004	0.002	0.003	0.002	0.003	0.004	0.003	0.003	0.003	0.004
120	0.005	0.003	0.004	0.003	0.003	0.004	0.003	0.004	0.004	0.004	0.005	0.005	0.002	0.004	0.003	0.005	0.005	0.004	0.003	0.005	0.004	0.004	0.005	0.004	0.002	0.003	0.005	0.002	0.003	0.002	0.003	0.005	0.003	0.003	0.004	0.005
135	0.007	0.003	0.005	0.003	0.004	0.005	0.004	0.005	0.005	0.005	0.007	0.007	0.003	0.004	0.003	0.008	0.007	0.004	0.003	0.006	0.004	0.005	0.007	0.005	0.003	0.004	0.006	0.003	0.003	0.003	0.003	0.006	0.003	0.003	0.005	0.006
150	0.01	0.004	0.006	0.003	0.005	0.007	0.005	0.006	0.007	0.006	0.01	0.011	0.003	0.005	0.003	0.011	0.01	0.006	0.004	0.008	0.006	0.006	0.01	0.007	0.003	0.005	0.008	0.003	0.004	0.003	0.004	0.008	0.004	0.004	0.007	0.008
165	0.015	0.004	0.008	0.003	0.006	0.009	0.006	0.009	0.009	0.008	0.014	0.016	0.003	0.007	0.004	0.017	0.014	0.007	0.004	0.012	0.007	0.008	0.015	0.01	0.003	0.006	0.012	0.003	0.004	0.003	0.004	0.012	0.005	0.004	0.01	0.012
180	0.022	0.005	0.01	0.004	0.008	0.013	0.008	0.012	0.013	0.01	0.022	0.025	0.003	0.009	0.004	0.026	0.022	0.01	0.005	0.017	0.01	0.011	0.023	0.014	0.003	0.008	0.018	0.003	0.005	0.003	0.005	0.017	0.006	0.005	0.014	0.018
195	0.035	0.006	0.014	0.004	0.01	0.019	0.01	0.017	0.019	0.014	0.034	0.039	0.004	0.013	0.005	0.042	0.035	0.014	0.007	0.026	0.013	0.015	0.035	0.02	0.003	0.011	0.027	0.003	0.006	0.004	0.007	0.026	0.007	0.006	0.02	0.027
210	0.055	0.007	0.02	0.005	0.014	0.028	0.014	0.025	0.028	0.02	0.053	0.062	0.004	0.018	0.005	0.067	0.054	0.019	0.008	0.04	0.019	0.022	0.056	0.029	0.004	0.015	0.041	0.004	0.007	0.004	0.008	0.04	0.009	0.007	0.029	0.041
225	0.087	0.009	0.029	0.005	0.019	0.042	0.019	0.037	0.042	0.029	0.082	0.098	0.004	0.025	0.006	0.105	0.085	0.027	0.01	0.061	0.026	0.031	0.087	0.044	0.004	0.021	0.062	0.004	0.009	0.004	0.01	0.061	0.012	0.008	0.044	0.062
240	0.132	0.011	0.041	0.006	0.026	0.061	0.026	0.054	0.061	0.041	0.125	0.149	0.005	0.035	0.007	0.16	0.129	0.038	0.013	0.091	0.037	0.045	0.133	0.065	0.005	0.028	0.093	0.005	0.011	0.005	0.013	0.091	0.015	0.01	0.065	0.093
255	0.193	0.014	0.059	0.007	0.035	0.089	0.036	0.077	0.089	0.059	0.184	0.219	0.006	0.049	0.009	0.233	0.189	0.054	0.017	0.134	0.052	0.064	0.195	0.094	0.005	0.039	0.137	0.005	0.014	0.006	0.017	0.134	0.019	0.013	0.094	0.137
270	0.272	0.018	0.083	0.009	0.048	0.127	0.049	0.11	0.127	0.083	0.258	0.305	0.007	0.068	0.011	0.324	0.266	0.075	0.022	0.191	0.073	0.091	0.274	0.134	0.006	0.054	0.196	0.006	0.018	0.006	0.022	0.191	0.025	0.016	0.134	0.196
285	0.363	0.022	0.114	0.01	0.066	0.176	0.067	0.152	0.176	0.114	0.347	0.403	0.007	0.094	0.013	0.425	0.356	0.104	0.028	0.261	0.101	0.126	0.366	0.185	0.007	0.074	0.267	0.007	0.022	0.007	0.028	0.261	0.033	0.021	0.185	0.267
300	0.462	0.028	0.155	0.012	0.088	0.237	0.09	0.205	0.237	0.155	0.444	0.505	0.009	0.127	0.016	0.529	0.454	0.141	0.036	0.343	0.136	0.17	0.464	0.248	0.008	0.099	0.351	0.008	0.028	0.008	0.036	0.343	0.043	0.026	0.248	0.351
315	0.56	0.036	0.204	0.014	0.116	0.307	0.118	0.269	0.307	0.204	0.541	0.604	0.01	0.168	0.019	0.627	0.552	0.186	0.046	0.432	0.18	0.224	0.563	0.322	0.009	0.132	0.441	0.009	0.036	0.01	0.046	0.432	0.055	0.032	0.322	0.441
330	0.651	0.045	0.263	0.017	0.151	0.385	0.154	0.34	0.385	0.263	0.633	0.693	0.011	0.217	0.023	0.714	0.643	0.24	0.058	0.523	0.233	0.286	0.654	0.401	0.01	0.171	0.531	0.01	0.045	0.011	0.058	0.523	0.071	0.041	0.401	0.531
345	0.73	0.056	0.328	0.02	0.193	0.466	0.196	0.417	0.466	0.328	0.714	0.767	0.013	0.274	0.028	0.786	0.723	0.302	0.074	0.609	0.293	0.356	0.733	0.484	0.011	0.218	0.617	0.011	0.056	0.013	0.074	0.609	0.09	0.051	0.484	0.617
360	0.796	0.07	0.399	0.024	0.241	0.547	0.245	0.495	0.547	0.399	0.782	0.828	0.015	0.337	0.034	0.843	0.79	0.369	0.092	0.687	0.359	0.429	0.798	0.565	0.013	0.271	0.695	0.013	0.07	0.015	0.092	0.687	0.113	0.063	0.565	0.695
375	0.849	0.086	0.471	0.028	0.295	0.623	0.3	0.571	0.623	0.471	0.837	0.874	0.018	0.404	0.041	0.887	0.843	0.439	0.115	0.754	0.428	0.503	0.85	0.64	0.015	0.33	0.761	0.015	0.086	0.017	0.115	0.754	0.14	0.077	0.64	0.761
390	0.889	0.105	0.543	0.034	0.354	0.692	0.359	0.643	0.692	0.543	0.879	0.91	0.02	0.473	0.049	0.919	0.885	0.51	0.141	0.81	0.498	0.576	0.89	0.708	0.017	0.392	0.816	0.017	0.105	0.02	0.141	0.81	0.172	0.095	0.708	0.816
405	0.92	0.128	0.611	0.04	0.415	0.752	0.421	0.707	0.752	0.611	0.912	0.936	0.023	0.54	0.058	0.943	0.916	0.578	0.171	0.855	0.566	0.643	0.921	0.766	0.019	0.456	0.86	0.019	0.128	0.023	0.171	0.855	0.208	0.115	0.766	0.86
420	0.942	0.154	0.674	0.047	0.477	0.803	0.483	0.763	0.803	0.674	0.936	0.955	0.027	0.605	0.069	0.96	0.94	0.642	0.205	0.891	0.631	0.704	0.943	0.816	0.022	0.52	0.895	0.022	0.154	0.026	0.205	0.891	0.248	0.138	0.816	0.895
435	0.959	0.183	0.73	0.055	0.539	0.845	0.545	0.81	0.845	0.73	0.954	0.969	0.031	0.665	0.082	0.973	0.957	0.7	0.243	0.918	0.689	0.758	0.96	0.856	0.025	0.582	0.922	0.025	0.183	0.03	0.243	0.918	0.292	0.165	0.856	0.922
450	0.971	0.216	0.779	0.064	0.598	0.879	0.604	0.85	0.879	0.779	0.968	0.979	0.036	0.72	0.097	0.982	0.97	0.752	0.284	0.94	0.742	0.804	0.972	0.889	0.029	0.64	0.942	0.029	0.216	0.035	0.284	0.94	0.339	0.195	0.889	0.942
465	0.981	0.252	0.82	0.075	0.653	0.907	0.659	0.882	0.907	0.82	0.978	0.986	0.041	0.767	0.114	0.989	0.979	0.797	0.328	0.956	0.788	0.842	0.981	0.914	0.033	0.694	0.958	0.033	0.252	0.04	0.328	0.956	0.389	0.227	0.914	0.958
480	0.988	0.29	0.855	0.087	0.703	0.928	0.709	0.908	0.928	0.855	0.985	0.993	0.047	0.809	0.133	0.995	0.987	0.835	0.374	0.968	0.827	0.874	0.988	0.935	0.037	0.742	0.969	0.037	0.29	0.046	0.374	0.968	0.439	0.263	0.935	0.969
495	0.994	0.331	0.884	0.101	0.749	0.945	0.754	0.928	0.945	0.884	0.992	0.998	0.054	0.844	0.154	1	0.993	0.866	0.421	0.977	0.86	0.9	0.994	0.95	0.043	0.784	0.978	0.043	0.331	0.052	0.421	0.977	0.489	0.301	0.95	0.978
510	0.999	0.374	0.908	0.116	0.789	0.959	0.793	0.945	0.959	0.908	0.997	1.003	0.062	0.873	0.177	1.005	0.998	0.893	0.469	0.984	0.887	0.922	0.999	0.963	0.048	0.82	0.985	0.048	0.374	0.059	0.469	0.984	0.539	0.341	0.963	0.985
525	1.003	0.417	0.927	0.133	0.824	0.969	0.828	0.958	0.969	0.927	1.002	1.008	0.071	0.898	0.203	1.01	1.003	0.914	0.516	0.99	0.909	0.939	1.004	0.972	0.055	0.852	0.991	0.055	0.417	0.068	0.516	0.99	0.587	0.383	0.972	0.991
540	1.008	0.461	0.943	0.151	0.853	0.977	0.857	0.968	0.977	0.943	1.006	1.013	0.08	0.918	0.23	1.016	1.007	0.932	0.563	0.995	0.928	0.952	1.008	0.98	0.062	0.878	0.996	0.062	0.461	0.077	0.563	0.995	0.632	0.425	0.98	0.996
555	1.013	0.505	0.955	0.172	0.879	0.984	0.882	0.976	0.984	0.955	1.011	1.019	0.091	0.934	0.26	1.022	1.012	0.946	0.607	1	0.943	0.963	1.014	0.986	0.07	0.901	1.001	0.07	0.505	0.087	0.607	1	0.675	0.468	0.986	1.001
570	1.019	0.548	0.965	0.194	0.9	0.989	0.903	0.983	0.989	0.965	1.017	1.025	0.102	0.948	0.291	1.029	1.018	0.958	0.649	1.005	0.955	0.972	1.02	0.992	0.079	0.919	1.006	0.079	0.548	0.098	0.649	1.005	0.714	0.511	0.992	1.006
585	1.026	0.59	0.973	0.218	0.918	0.994	0.921	0.988	0.994	0.973	1.023	1.033	0.115	0.959	0.323	1.038	1.025	0.967	0.688	1.009	0.965	0.979	1.026	0.996	0.089	0.934	1.01	0.089	0.59	0.11	0.688	1.009	0.75	0.552	0.996	1.01
600	1.034	0.63	0.98	0.243	0.933	0.999	0.935	0.993	0.999	0.98	1.03	1.042	0.129	0.968	0.357	1.048	1.032	0.975	0.725	1.014	0.973	0.985	1.034	1.001	0.099	0.947	1.015	0.099	0.63	0.124	0.725	1.014	0.783	0.592	1.001	1.015
615	1.042	0.667	0.986	0.27	0.946	1.003	0.947	0.998	1.003	0.986	1.038	1.052	0.144	0.975	0.392	1.059	1.041	0.981	0.758	1.02	0.979	0.99	1.043	1.005	0.111	0.957	1.021	0.111	0.667	0.138	0.758	1.02	0.812	0.631	1.005	1.021
630	1.052	0.702	0.991	0.298	0.956	1.008	0.957	1.002	1.008	0.991	1.048	1.064	0.161	0.981	0.427	1.071	1.05	0.986	0.788	1.026	0.985	0.995	1.053	1.01	0.123	0.966	1.028	0.123	0.702	0.154	0.788	1.026	0.838	0.667	1.01	1.028
645	1.063	0.735	0.995	0.327	0.965	1.012	0.966	1.006	1.012	0.995	1.058	1.076	0.178	0.986	0.462	1.084	1.061	0.991	0.815	1.033	0.99	0.999	1.064	1.014	0.137	0.973	1.035	0.137	0.735	0.171	0.815	1.033	0.861	0.702	1.014	1.035
660	1.075	0.765	0.999	0.357	0.972	1.017	0.973	1.011	1.017	0.999	1.069	1.089	0.197	0.991	0.497	1.097	1.073	0.996	0.839	1.041	0.994	1.003	1.076	1.02	0.152	0.979	1.043	0.152	0.765	0.189	0.839	1.041	0.881	0.733	1.02	1.043
675	1.088	0.792	1.003	0.387	0.978	1.023	0.979	1.015	1.023	1.003	1.082	1.103	0.217	0.995	0.532	1.111	1.085	1	0.861	1.05	0.998	1.007	1.089	1.025	0.167	0.985	1.052	0.167	0.792	0.208	0.861	1.05	0.898	0.763	1.025	1.052
690	1.101	0.817	1.007	0.419	0.983	1.029	0.984	1.021	1.029	1.007	1.094	1.116	0.238	0.999	0.566	1.124	1.098	1.004	0.88	1.06	1.002	1.011	1.102	1.032	0.184	0.989	1.063	0.184	0.817	0.229	0.88	1.06	0.913	0.789	1.032	1.063
705	1.114	0.839	1.012	0.45	0.988	1.036	0.989	1.026	1.036	1.012	1.107	1.129	0.26	1.003	0.599	1.137	1.111	1.008	0.896	1.071	1.006	1.016	1.115	1.039	0.202	0.993	1.073	0.202	0.839	0.25	0.896	1.071	0.926	0.814	1.039	1.073
720	1.126	0.859	1.016	0.481	0.992	1.043	0.993	1.033	1.043	1.016	1.12	1.141	0.283	1.007	0.63	1.148	1.123	1.012	0.911	1.082	1.011	1.021	1.127	1.047	0.221	0.997	1.085	0.221	0.859	0.272	0.911	1.082	0.937	0.836	1.047	1.085
735	1.138	0.877	1.022	0.512	0.996	1.051	0.997	1.04	1.051	1.022	1.132	1.152	0.307	1.012	0.66	1.158	1.135	1.017	0.924	1.094	1.015	1.027	1.139	1.056	0.24	1.001	1.097	0.24	0.877	0.296	0.924	1.094	0.947	0.855	1.056	1.097
750	1.149	0.893	1.027	0.543	1	1.061	1.001	1.047	1.061	1.027	1.143	1.161	0.332	1.016	0.689	1.167	1.146	1.022	0.935	1.106	1.02	1.033	1.15	1.066	0.261	1.005	1.109	0.261	0.893	0.319	0.935	1.106	0.955	0.873	1.066	1.109
765	1.159	0.907	1.033	0.573	1.004	1.07	1.005	1.056	1.07	1.033	1.153	1.17	0.357	1.021	0.716	1.175	1.156	1.027	0.944	1.118	1.025	1.04	1.159	1.076	0.282	1.009	1.121	0.282	0.907	0.344	0.944	1.118	0.963	0.889	1.076	1.121
780	1.167	0.919	1.04	0.602	1.008	1.08	1.008	1.065	1.08	1.04	1.163	1.177	0.383	1.026	0.742	1.181	1.165	1.033	0.952	1.129	1.031	1.047	1.168	1.086	0.304	1.013	1.132	0.304	0.919	0.369	0.952	1.129	0.969	0.903	1.086	1.132
795	1.175	0.93	1.047	0.63	1.012	1.091	1.012	1.074	1.091	1.047	1.171	1.183	0.409	1.032	0.766	1.187	1.173	1.04	0.96	1.14	1.037	1.055	1.175	1.097	0.327	1.018	1.142	0.327	0.93	0.395	0.96	1.14	0.974	0.915	1.097	1.142
810	1.181	0.939	1.055	0.657	1.016	1.102	1.017	1.085	1.102	1.055	1.177	1.188	0.435	1.038	0.788	1.191	1.179	1.047	0.966	1.15	1.044	1.064	1.181	1.108	0.35	1.023	1.152	0.35	0.939	0.421	0.966	1.15	0.979	0.926	1.108	1.152
825	1.186	0.947	1.063	0.683	1.021	1.113	1.021	1.095	1.113	1.063	1.183	1.192	0.461	1.045	0.808	1.194	1.185	1.054	0.972	1.159	1.051	1.073	1.186	1.119	0.374	1.028	1.161	0.374	0.947	0.447	0.972	1.159	0.984	0.936	1.119	1.161
840	1.19	0.955	1.072	0.707	1.025	1.123	1.026	1.105	1.123	1.072	1.188	1.195	0.487	1.053	0.827	1.197	1.189	1.063	0.976	1.166	1.059	1.083	1.191	1.129	0.398	1.033	1.169	0.398	0.955	0.472	0.976	1.166	0.988	0.944	1.129	1.169
855	1.194	0.961	1.082	0.73	1.031	1.133	1.032	1.116	1.133	1.082	1.192	1.198	0.513	1.06	0.844	1.199	1.193	1.071	0.981	1.173	1.068	1.092	1.194	1.139	0.423	1.04	1.175	0.423	0.961	0.498	0.981	1.173	0.991	0.952	1.139	1.175
870	1.197	0.967	1.091	0.752	1.036	1.143	1.038	1.126	1.143	1.091	1.195	1.2	0.539	1.069	0.86	1.201	1.196	1.08	0.985	1.179	1.077	1.102	1.197	1.148	0.447	1.046	1.181	0.447	0.967	0.524	0.985	1.179	0.995	0.958	1.148	1.181
885	1.199	0.972	1.101	0.773	1.043	1.152	1.044	1.135	1.152	1.101	1.197	1.201	0.564	1.078	0.874	1.202	1.198	1.09	0.989	1.184	1.086	1.112	1.199	1.157	0.471	1.053	1.186	0.471	0.972	0.549	0.989	1.184	0.998	0.964	1.157	1.186
900	1.201	0.977	1.111	0.792	1.049	1.16	1.051	1.144	1.16	1.111	1.199	1.203	0.589	1.087	0.887	1.203	1.2	1.099	0.992	1.189	1.095	1.122	1.201	1.164	0.496	1.061	1.19	0.496	0.977	0.574	0.992	1.189	1.002	0.969	1.164	1.19
915	1.202	0.981	1.12	0.81	1.056	1.167	1.058	1.153	1.167	1.12	1.201	1.204	0.613	1.096	0.899	1.204	1.202	1.109	0.996	1.192	1.105	1.131	1.202	1.171	0.52	1.069	1.193	0.52	0.981	0.598	0.996	1.192	1.005	0.974	1.171	1.193
930	1.203	0.985	1.13	0.827	1.064	1.173	1.066	1.16	1.173	1.13	1.202	1.204	0.636	1.105	0.909	1.205	1.203	1.118	0.999	1.195	1.114	1.14	1.203	1.177	0.543	1.077	1.196	0.543	0.985	0.621	0.999	1.195	1.009	0.978	1.177	1.196
945	1.204	0.988	1.138	0.842	1.072	1.179	1.074	1.167	1.179	1.138	1.203	1.205	0.658	1.114	0.919	1.205	1.204	1.127	1.002	1.198	1.123	1.149	1.204	1.182	0.567	1.086	1.198	0.567	0.988	0.644	1.002	1.198	1.012	0.982	1.182	1.198
960	1.205	0.992	1.147	0.856	1.08	1.183	1.082	1.173	1.183	1.147	1.204	1.205	0.68	1.124	0.928	1.206	1.205	1.136	1.006	1.2	1.132	1.156	1.205	1.186	0.59	1.094	1.2	0.59	0.992	0.666	1.006	1.2	1.016	0.986	1.186	1.2
975	1.205	0.995	1.154	0.869	1.088	1.188	1.09	1.179	1.188	1.154	1.205	1.206	0.701	1.132	0.936	1.206	1.205	1.144	1.009	1.201	1.141	1.163	1.205	1.19	0.612	1.103	1.202	0.612	0.995	0.687	1.009	1.201	1.02	0.989	1.19	1.202
990	1.206	0.998	1.161	0.881	1.097	1.191	1.099	1.183	1.191	1.161	1.205	1.206	0.721	1.141	0.943	1.206	1.206	1.152	1.013	1.203	1.149	1.17	1.206	1.193	0.633	1.112	1.203	0.633	0.998	0.707	1.013	1.203	1.025	0.993	1.193	1.203
1005	1.206	1.001	1.168	0.892	1.105	1.194	1.107	1.188	1.194	1.168	1.206	1.206	0.74	1.148	0.949	1.206	1.206	1.159	1.017	1.204	1.156	1.175	1.206	1.196	0.654	1.12	1.204	0.654	1.001	0.726	1.017	1.204	1.029	0.996	1.196	1.204
1020	1.206	1.005	1.174	0.902	1.114	1.197	1.116	1.191	1.197	1.174	1.206	1.206	0.757	1.156	0.955	1.207	1.206	1.166	1.021	1.204	1.163	1.18	1.206	1.198	0.675	1.129	1.205	0.675	1.005	0.745	1.021	1.204	1.034	0.999	1.198	1.205
1035	1.207	1.008	1.179	0.912	1.122	1.199	1.124	1.194	1.199	1.179	1.206	1.207	0.774	1.162	0.96	1.207	1.207	1.172	1.025	1.205	1.169	1.185	1.207	1.2	0.694	1.137	1.205	0.694	1.008	0.762	1.025	1.205	1.039	1.002	1.2	1.205
1050	1.207	1.011	1.183	0.92	1.13	1.2	1.132	1.196	1.2	1.183	1.207	1.207	0.791	1.168	0.965	1.207	1.207	1.177	1.029	1.206	1.174	1.188	1.207	1.202	0.713	1.145	1.206	0.713	1.011	0.779	1.029	1.206	1.045	1.006	1.202	1.206
1065	1.207	1.015	1.187	0.928	1.138	1.202	1.14	1.199	1.202	1.187	1.207	1.207	0.806	1.174	0.969	1.207	1.207	1.181	1.034	1.206	1.179	1.192	1.207	1.203	0.731	1.152	1.206	0.731	1.015	0.794	1.034	1.206	1.051	1.009	1.203	1.206
1080	1.207	1.019	1.19	0.935	1.146	1.203	1.147	1.2	1.203	1.19	1.207	1.207	0.82	1.179	0.973	1.207	1.207	1.185	1.039	1.207	1.183	1.194	1.207	1.204	0.748	1.158	1.207	0.748	1.019	0.809	1.039	1.207	1.057	1.012	1.204	1.207
1095	1.207	1.022	1.193	0.941	1.152	1.204	1.154	1.202	1.204	1.193	1.207	1.207	0.833	1.183	0.977	1.207	1.207	1.189	1.044	1.207	1.187	1.197	1.207	1.205	0.764	1.164	1.207	0.764	1.022	0.823	1.044	1.207	1.063	1.016	1.205	1.207
1110	1.207	1.027	1.196	0.947	1.159	1.205	1.16	1.203	1.205	1.196	1.207	1.207	0.846	1.187	0.981	1.207	1.207	1.192	1.05	1.207	1.191	1.199	1.207	1.205	0.78	1.17	1.207	0.78	1.027	0.836	1.05	1.207	1.07	1.019	1.205	1.207
1125	1.207	1.031	1.198	0.952	1.165	1.206	1.166	1.204	1.206	1.198	1.207	1.207	0.858	1.19	0.984	1.207	1.207	1.195	1.056	1.207	1.193	1.201	1.207	1.206	0.794	1.175	1.207	0.794	1.031	0.848	1.056	1.207	1.077	1.023	1.206	1.207
1140	1.207	1.035	1.2	0.957	1.17	1.206	1.172	1.205	1.206	1.2	1.207	1.207	0.868	1.193	0.987	1.207	1.207	1.197	1.062	1.207	1.196	1.202	1.207	1.207	0.808	1.18	1.207	0.808	1.035	0.86	1.062	1.207	1.084	1.027	1.207	1.207
1155	1.207	1.04	1.201	0.962	1.175	1.207	1.176	1.206	1.207	1.201	1.207	1.207	0.879	1.196	0.99	1.207	1.207	1.199	1.068	1.207	1.198	1.203	1.207	1.207	0.821	1.184	1.207	0.821	1.04	0.87	1.068	1.207	1.091	1.032	1.207	1.207
1170	1.207	1.045	1.203	0.966	1.18	1.207	1.181	1.206	1.207	1.203	1.207	1.207	0.888	1.198	0.993	1.207	1.207	1.201	1.074	1.208	1.2	1.204	1.207	1.207	0.834	1.187	1.208	0.834	1.045	0.88	1.074	1.208	1.098	1.036	1.207	1.208
1185	1.207	1.051	1.204	0.97	1.184	1.207	1.185	1.207	1.207	1.204	1.207	1.207	0.897	1.2	0.996	1.207	1.207	1.202	1.081	1.208	1.201	1.205	1.207	1.207	0.845	1.19	1.208	0.845	1.051	0.89	1.081	1.208	1.105	1.041	1.207	1.208
1200	1.207	1.056	1.205	0.974	1.187	1.208	1.188	1.207	1.208	1.205	1.207	1.207	0.905	1.201	0.999	1.207	1.207	1.203	1.088	1.208	1.203	1.206	1.207	1.208	0.856	1.193	1.208	0.856	1.056	0.898	1.088	1.208	1.112	1.046	1.208	1.208
1215	1.207	1.062	1.206	0.977	1.19	1.208	1.191	1.207	1.208	1.206	1.207	1.207	0.913	1.203	1.002	1.207	1.207	1.204	1.095	1.208	1.204	1.206	1.207	1.208	0.867	1.196	1.208	0.867	1.062	0.906	1.095	1.208	1.119	1.051	1.208	1.208
1230	1.207	1.068	1.206	0.98	1.193	1.208	1.194	1.208	1.208	1.206	1.207	1.207	0.92	1.204	1.005	1.207	1.207	1.205	1.102	1.208	1.205	1.207	1.207	1.208	0.876	1.198	1.208	0.876	1.068	0.914	1.102	1.208	1.126	1.057	1.208	1.208
1245	1.207	1.074	1.207	0.983	1.195	1.208	1.196	1.208	1.208	1.207	1.207	1.207	0.926	1.205	1.008	1.207	1.207	1.206	1.109	1.208	1.206	1.207	1.207	1.208	0.885	1.2	1.208	0.885	1.074	0.921	1.109	1.208	1.133	1.062	1.208	1.208
1260	1.207	1.08	1.207	0.986	1.197	1.208	1.198	1.208	1.208	1.207	1.207	1.207	0.932	1.206	1.011	1.207	1.207	1.207	1.115	1.208	1.206	1.208	1.207	1.208	0.894	1.201	1.208	0.894	1.08	0.927	1.115	1.208	1.139	1.068	1.208	1.208
1275	1.207	1.086	1.208	0.989	1.199	1.208	1.2	1.208	1.208	1.208	1.207	1.207	0.938	1.206	1.014	1.207	1.207	1.207	1.122	1.208	1.207	1.208	1.207	1.208	0.902	1.203	1.208	0.902	1.086	0.933	1.122	1.208	1.146	1.074	1.208	1.208
1290	1.207	1.093	1.208	0.992	1.201	1.208	1.201	1.208	1.208	1.208	1.207	1.207	0.943	1.207	1.017	1.207	1.207	1.208	1.129	1.208	1.207	1.208	1.207	1.208	0.909	1.204	1.208	0.909	1.093	0.938	1.129	1.208	1.152	1.08	1.208	1.208
1305	1.207	1.099	1.208	0.994	1.202	1.208	1.203	1.209	1.208	1.208	1.207	1.207	0.948	1.207	1.02	1.207	1.207	1.208	1.135	1.208	1.208	1.208	1.207	1.208	0.916	1.205	1.208	0.916	1.099	0.944	1.135	1.208	1.157	1.086	1.208	1.208
1320	1.207	1.106	1.209	0.997	1.204	1.209	1.204	1.209	1.209	1.209	1.207	1.207	0.953	1.208	1.024	1.207	1.207	1.208	1.141	1.208	1.208	1.209	1.207	1.208	0.922	1.206	1.208	0.922	1.106	0.948	1.141	1.208	1.162	1.093	1.208	1.208
1335	1.207	1.112	1.209	1	1.205	1.209	1.205	1.209	1.209	1.209	1.207	1.207	0.957	1.208	1.028	1.207	1.207	1.209	1.147	1.208	1.208	1.209	1.207	1.209	0.928	1.206	1.208	0.928	1.112	0.953	1.147	1.208	1.167	1.099	1.209	1.208
1350	1.207	1.119	1.209	1.003	1.206	1.209	1.206	1.209	1.209	1.209	1.207	1.207	0.961	1.208	1.031	1.207	1.207	1.209	1.153	1.208	1.209	1.209	1.207	1.209	0.933	1.207	1.208	0.933	1.119	0.957	1.153	1.208	1.172	1.105	1.209	1.208
1365	1.207	1.125	1.209	1.005	1.206	1.209	1.207	1.209	1.209	1.209	1.207	1.207	0.964	1.209	1.035	1.207	1.207	1.209	1.158	1.208	1.209	1.209	1.207	1.209	0.939	1.208	1.208	0.939	1.125	0.961	1.158	1.208	1.176	1.112	1.209	1.208
1380	1.207	1.131	1.209	1.008	1.207	1.209	1.207	1.209	1.209	1.209	1.207	1.207	0.968	1.209	1.04	1.207	1.207	1.209	1.163	1.208	1.209	1.209	1.207	1.209	0.943	1.208	1.208	0.943	1.131	0.964	1.163	1.208	1.18	1.118	1.209	1.208
1395	1.207	1.137	1.209	1.011	1.208	1.209	1.208	1.209	1.209	1.209	1.207	1.207	0.971	1.209	1.044	1.207	1.207	1.209	1.168	1.208	1.209	1.209	1.207	1.209	0.948	1.208	1.208	0.948	1.137	0.968	1.168	1.208	1.183	1.124	1.209	1.208
1410	1.207	1.143	1.209	1.014	1.208	1.209	1.208	1.209	1.209	1.209	1.207	1.207	0.974	1.209	1.048	1.207	1.207	1.209	1.172	1.208	1.209	1.209	1.207	1.209	0.952	1.209	1.208	0.952	1.143	0.971	1.172	1.208	1.186	1.13	1.209	1.208
1425	1.207	1.148	1.209	1.017	1.209	1.209	1.209	1.209	1.209	1.209	1.207	1.207	0.977	1.209	1.053	1.207	1.207	1.209	1.176	1.208	1.209	1.209	1.207	1.209	0.956	1.209	1.208	0.956	1.148	0.974	1.176	1.208	1.189	1.136	1.209	1.208
1440	1.207	1.153	1.209	1.02	1.209	1.209	1.209	1.209	1.209	1.209	1.207	1.207	0.98	1.21	1.058	1.207	1.207	1.21	1.18	1.208	1.21	1.209	1.207	1.209	0.96	1.209	1.208	0.96	1.153	0.977	1.18	1.208	1.192	1.141	1.209	1.208
"""

# Define the mapping based on the user's description
mapping = {
    'A1': {'strain': 'Strain1', 'type': 'WT', 'rep': 'Rep1', 'cond': 'A'},
    'A2': {'strain': 'Strain1', 'type': 'MUT', 'rep': 'Rep1', 'cond': 'A'},
    'A3': {'strain': 'Strain1', 'type': 'WT', 'rep': 'Rep2', 'cond': 'A'},
    'A4': {'strain': 'Strain1', 'type': 'MUT', 'rep': 'Rep2', 'cond': 'A'},
    'A5': {'strain': 'Strain2', 'type': 'WT', 'rep': 'Rep1', 'cond': 'A'},
    'A6': {'strain': 'Strain2', 'type': 'MUT', 'rep': 'Rep1', 'cond': 'A'},
    'A7': {'strain': 'Strain2', 'type': 'WT', 'rep': 'Rep2', 'cond': 'A'},
    'A8': {'strain': 'Strain2', 'type': 'MUT', 'rep': 'Rep2', 'cond': 'A'},
    'A9': {'strain': 'Strain3', 'type': 'WT', 'rep': 'Rep1', 'cond': 'A'},
    'A10': {'strain': 'Strain3', 'type': 'MUT', 'rep': 'Rep1', 'cond': 'A'},
    'A11': {'strain': 'Strain3', 'type': 'WT', 'rep': 'Rep2', 'cond': 'A'},
    'A12': {'strain': 'Strain3', 'type': 'MUT', 'rep': 'Rep2', 'cond': 'A'},
    'B1': {'strain': 'Strain1', 'type': 'WT', 'rep': 'Rep1', 'cond': 'B'},
    'B2': {'strain': 'Strain1', 'type': 'MUT', 'rep': 'Rep1', 'cond': 'B'},
    'B3': {'strain': 'Strain1', 'type': 'WT', 'rep': 'Rep2', 'cond': 'B'},
    'B4': {'strain': 'Strain1', 'type': 'MUT', 'rep': 'Rep2', 'cond': 'B'},
    'B5': {'strain': 'Strain2', 'type': 'WT', 'rep': 'Rep1', 'cond': 'B'},
    'B6': {'strain': 'Strain2', 'type': 'MUT', 'rep': 'Rep1', 'cond': 'B'},
    'B7': {'strain': 'Strain2', 'type': 'WT', 'rep': 'Rep2', 'cond': 'B'},
    'B8': {'strain': 'Strain2', 'type': 'MUT', 'rep': 'Rep2', 'cond': 'B'},
    'B9': {'strain': 'Strain3', 'type': 'WT', 'rep': 'Rep1', 'cond': 'B'},
    'B10': {'strain': 'Strain3', 'type': 'MUT', 'rep': 'Rep1', 'cond': 'B'},
    'B11': {'strain': 'Strain3', 'type': 'WT', 'rep': 'Rep2', 'cond': 'B'},
    'B12': {'strain': 'Strain3', 'type': 'MUT', 'rep': 'Rep2', 'cond': 'B'},
    'C1': {'strain': 'Strain1', 'type': 'WT', 'rep': 'Rep1', 'cond': 'C'},
    'C2': {'strain': 'Strain1', 'type': 'MUT', 'rep': 'Rep1', 'cond': 'C'},
    'C3': {'strain': 'Strain1', 'type': 'WT', 'rep': 'Rep2', 'cond': 'C'},
    'C4': {'strain': 'Strain1', 'type': 'MUT', 'rep': 'Rep2', 'cond': 'C'},
    'C5': {'strain': 'Strain2', 'type': 'WT', 'rep': 'Rep1', 'cond': 'C'},
    'C6': {'strain': 'Strain2', 'type': 'MUT', 'rep': 'Rep1', 'cond': 'C'},
    'C7': {'strain': 'Strain2', 'type': 'WT', 'rep': 'Rep2', 'cond': 'C'},
    'C8': {'strain': 'Strain2', 'type': 'MUT', 'rep': 'Rep2', 'cond': 'C'},
    'C9': {'strain': 'Strain3', 'type': 'WT', 'rep': 'Rep1', 'cond': 'C'},
    'C10': {'strain': 'Strain3', 'type': 'MUT', 'rep': 'Rep1', 'cond': 'C'},
    'C11': {'strain': 'Strain3', 'type': 'WT', 'rep': 'Rep2', 'cond': 'C'},
    'C12': {'strain': 'Strain3', 'type': 'MUT', 'rep': 'Rep2', 'cond': 'C'},
}

# Load the data
try:
    data = pd.read_csv(io.StringIO(data_string), sep='\t')
except pd.errors.ParserError as e:
    print(f"Error parsing data. Please check the format. Details:\n{e}")
    raise e  # Raise exception
except Exception as e:
    print(f"An unexpected error occurred during data loading: {e}")
    raise e  # Raise exception

# --- Run Analysis (without showing individual plots) ---
# Set show_plots=False to suppress the individual growth curve plots
analysis_results = analyze_growth_curves(
    data, mapping, k_threshold_percent=0.95, show_plots=False)

# --- Generate Box Plot ---
if analysis_results is not None:
    print("\nGenerating WT vs MUT Time-to-K box plot...")
    try:
        # Ensure 'Time_to_K_hours' and 'Type' columns exist
        if 'Time_to_K_hours' not in analysis_results.columns or 'Type' not in analysis_results.columns:
            raise KeyError(
                "Required columns ('Time_to_K_hours', 'Type') not found in analysis results.")

        # Drop rows where Time_to_K_hours is missing
        plot_data = analysis_results.dropna(subset=['Time_to_K_hours'])

        if plot_data.empty:
            print(
                "Warning: No valid Time-to-K data found after processing. Cannot generate box plot.")
        else:
            # Create the box plot using Seaborn (simplified call)
            plt.figure(figsize=(6, 7))  # Adjust figure size for box plot

            # Removed the palette argument to troubleshoot the 'boxprops' error
            sns.boxplot(data=plot_data, x='Type', y='Time_to_K_hours')

            # Add labels and title
            plt.xlabel('Strain Type')
            plt.ylabel('Time to 95% K (hours)')
            plt.title('Comparison of Time to Reach Carrying Capacity')
            # Grid lines on y-axis only
            plt.grid(True, linestyle='--', alpha=0.6, axis='y')
            plt.tight_layout()
            plt.show()  # Display the box plot

    except KeyError as e:
        print(
            f"Error during plotting: Missing column {e}. Check 'analysis_results' DataFrame.")
        print("Analysis Results Head:")
        # Use try-except for tabulate
        try:
            # Ensure tabulate is installed: pip install tabulate
            print(analysis_results.head().to_markdown(index=False))
        except ImportError:
            print(analysis_results.head())

    except Exception as e:
        print(f"An unexpected error occurred during box plot generation: {e}")

# --- Perform T-test ---
if analysis_results is not None:
    print("\n--- Statistical Test (T-test) ---")
    try:
        # Prepare data for t-test (drop NaNs)
        t_test_data = analysis_results.dropna(subset=['Time_to_K_hours'])

        # Separate the times for WT and MUT groups
        wt_times = t_test_data[t_test_data['Type'] == 'WT']['Time_to_K_hours']
        mut_times = t_test_data[t_test_data['Type']
                                == 'MUT']['Time_to_K_hours']

        if len(wt_times) < 2 or len(mut_times) < 2:
            print(
                "Warning: Not enough data points in one or both groups to perform t-test.")
        else:
            # Perform independent samples t-test
            # equal_var=False performs Welch's t-test, which doesn't assume equal variances
            t_stat, p_value = stats.ttest_ind(
                wt_times, mut_times, equal_var=False)

            print(f"Independent Samples T-test Results (WT vs MUT Time_to_K_hours):")
            print(f"T-statistic: {t_stat:.4f}")
            print(f"P-value: {p_value:.4f}")

            # Interpretation
            alpha = 0.05
            if p_value < alpha:
                print(
                    f"\nConclusion: Since the p-value ({p_value:.4f}) is less than {alpha}, we reject the null hypothesis.")
                print("There is a statistically significant difference in the time to reach carrying capacity between WT and MUT strains.")
            else:
                print(
                    f"\nConclusion: Since the p-value ({p_value:.4f}) is greater than or equal to {alpha}, we fail to reject the null hypothesis.")
                print("There is not enough statistical evidence to conclude a significant difference in the time to reach carrying capacity between WT and MUT strains.")

    except KeyError as e:
        print(
            f"Error during t-test: Missing column {e}. Check 'analysis_results' DataFrame.")
    except Exception as e:
        print(f"An unexpected error occurred during the t-test: {e}")


# Display the results table (optional, can be commented out if not needed)
# if analysis_results is not None:
#     print("\n--- Time to Reach 95% Carrying Capacity (K) ---")
#     try:
#         # Ensure tabulate is installed: pip install tabulate
#         print(analysis_results.round(3).to_markdown(index=False, numalign="left", stralign="left"))
#     except ImportError:
#         print("Info: 'tabulate' package not installed. Printing DataFrame without markdown formatting.")
#         print(analysis_results.round(3))
#     except Exception as e:
#         print(f"An error occurred during results display: {e}")


# Solution
# Independent Samples T-test Results (WT vs MUT Time_to_K_hours):
# T-statistic: 0.5945
# P-value: 0.5561

# Conclusion: Since the p-value (0.5561) is greater than or equal to 0.05, we fail to reject the null hypothesis.
# There is not enough statistical evidence to conclude a significant difference in the time to reach carrying capacity between WT and MUT strains.
```
